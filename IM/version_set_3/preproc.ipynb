```python
import pandas as pd
df = pd.read_csv('IM_new.csv')
df
```

```python
df.dtypes
```

```python
df_columns = []
division_for_columns = []
for column in df.columns[9:]:
    df[column].fillna(0, inplace=True)
    df[column] = df[column].apply(lambda x: float(str(x).replace(',', '.')))
    div = df[column].max()*1.3
    df[column]=df[column]/div
    df[column]=df[column].replace(0, df[column].mean())#fillna(df[column].mean(), inplace=True)
    df_columns.append(column)
    division_for_columns.append(div)
```

```python
order = pd.DataFrame((df_columns, division_for_columns)).T
order.columns = ('Columns', 'Division_for_normalization')
order
```

```python
df
```

```python
order.to_csv('Division_for_normalization.csv', index=False)
df.to_csv('IM_after_preproc.csv', index=False)
```
